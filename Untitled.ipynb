```python
fig = plt.figure() # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.4

df.query("y == 'no'").groupby('job').size().sort_values(ascending = False).plot(kind='bar', color='red', ax=ax, width=width, position=1)
df.query("y == 'yes'").groupby('job').size().plot(kind='bar', color='blue', ax=ax2, width=width, position=0)

ax.set_ylabel('NÃ£o Inscritos', color = 'red')
ax2.set_ylabel('Inscritos', color = 'blue')
##plt.legend()
plt.show()
```
