```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv('data/raw/bank-full.csv', sep=";")
df.head()
```

```python
df[['job','age']]
```

```python
for i in df.columns[1:4]:
    if type(i) !='int64':
        fig = plt.figure() # Create matplotlib figure
        ax = fig.add_subplot(111) # Create matplotlib axes
        ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.
        width = 0.4
    
        df.query("y == 'no'").groupby(i).size().sort_values(ascending = False).plot(kind='bar', color='red', ax=ax, width=width, position=1)
        df.query("y == 'yes'").groupby(i).size().plot(kind='bar', color='blue', ax=ax2, width=width, position=0)


        ax.set_ylabel('NÃ£o Inscritos', color = 'red')
        ax2.set_ylabel('Inscritos', color = 'blue')
        ##plt.legend()
        plt.show()
        
```

```python
df_no_subs = df.query("y == 'no'")
```

```python
df_subs = df.query("y == 'yes'")
```

```python
plt.figure(figsize=(10, 5))
plt.title("Histograma de idade - Inscritos")
sns.histplot(x=df_subs['age'], label='Age')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title("Histograma de idade - NÃ£o Inscritos")
sns.histplot(x=df_no_subs['age'], label='Age')
plt.show()
```

```python
plt.title("Histograma de idade")
plt.hist(df_no_subs.age, label='NÃ£o Inscritos', alpha=0.4, color='red')
plt.hist(df_subs.age,label='Inscritos', alpha=1, color='blue') 
plt.legend()
plt.show()
```
